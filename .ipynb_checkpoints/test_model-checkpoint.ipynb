```python
import zipfile

# ZIP 파일 경로
zip_file_path = r"D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data.zip"

# 압축을 해제할 대상 경로
extract_to_path = r"D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data"

# 경로 생성 (존재하지 않으면 생성)
os.makedirs(extract_to_path, exist_ok=True)

# ZIP 파일 열고 압축 해제
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_path)

print(f"압축이 해제되었습니다: {extract_to_path}")
```

```python
import yaml

# 데이터 정의
data = {
  'train': r'D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data\train\images\\',
  'val': r'D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data\valid\images\\',
  'test': r'D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data\test\images\\',
  'names': ['cat', 'dog'],
  'nc': 2
}

# YAML 파일 작성
yaml_file_path = r'D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data\Dog_Cat_Data.yaml'

with open(yaml_file_path, 'w') as f:
    yaml.dump(data, f)

# YAML 파일 읽기 및 출력
with open(yaml_file_path, 'r') as f:
    print(yaml.safe_load(f))
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
import os

# YOLOv8 학습 실행
results = model.train(
    data=r'D:\Python_Pytorch_Workspace\Jupyter_Test\Dog_Cat_Data\Dog_Cat_Data.yaml',  # 데이터 경로
    epochs=100,
    patience=30,
    batch=32,
    imgsz=416,
    project=r'D:\Python_Pytorch_Workspace\Jupyter_Test',  # 프로젝트 경로 지정
    name='modelresult'  # 실험 이름 지정
)
# 결과 확인
if results is not None and hasattr(results, 'save_dir'):
    default_save_path = results.save_dir
    print(f"저장된 경로: {default_save_path}")
else:
    # 결과가 None인 경우 기본 경로 확인
    default_save_path = os.path.join(r'D:\Python_Pytorch_Workspace\Jupyter_Test', 'modelresult')
    print(f"결과 객체가 None이므로 기본 경로를 사용합니다: {default_save_path}")


# 저장된 경로 확인
# default_save_path = results.save_dir

# # 디렉토리 확인 및 가장 최근 실험 선택
# if os.path.exists(default_save_path) and os.listdir(default_save_path):
#     experiment_dirs = sorted(os.listdir(default_save_path))
#     experiment_name = experiment_dirs[-1]
#     experiment_path = os.path.join(default_save_path, experiment_name)
#     print(f"가장 최근 실험 디렉토리 경로: {experiment_path}")
# else:
#     print(f"'{default_save_path}' 경로에 실험 디렉토리가 없습니다.")
```
